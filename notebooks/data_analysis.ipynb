```python
import dask.dataframe as dd
import pandas as pd
import plotly.express as px
```

```python
df_ibge = pd.read_csv('../data/DADOS.txt')
df_ibge
```

```python
df_filtered = df_ibge[['MUNICÍPIO', 'PRIMEIRO_NOME']].copy()
df_filtered
```

```python
df_filtered.info()
```

```python
df_filtered[df_filtered['PRIMEIRO_NOME'].isnull()]
```

```python
df_filtered.dropna(subset=['PRIMEIRO_NOME'], inplace=True)
df_filtered.info()
```

```python
df_filtered['PRIMEIRO_NOME'].sort_values().unique().tolist()
```

```python
df_filtered['MUNICÍPIO'].value_counts()
```

```python
df_filtered['MUNICÍPIO'] = df_filtered['MUNICÍPIO'].apply(lambda x: x.strip())
df_filtered['PRIMEIRO_NOME'] = df_filtered['PRIMEIRO_NOME'].apply(
    lambda x: ' '.join([w for w in x.strip().split() if len(w) >= 3])
)
df_filtered
```

```python
df_municipios = pd.read_csv('../data/populacao_municipios/Censo 2022 - População residente - Municípios.csv', sep=';')
df_municipios
```

```python
df_estados_nordeste = df_municipios.query(f'UF in {["MA", "PI", "CE", "RN", "PB", "PE", "AL", "SE", "BA"]}')
df_estados_nordeste['Municípios'] = df_estados_nordeste['Municípios'].apply(lambda x: x.strip().upper())
df_estados_nordeste = df_estados_nordeste.rename(columns={'Municípios': 'MUNICÍPIO'})
df_estados_nordeste
```

```python
df_merged = df_estados_nordeste.merge(df_filtered, how='inner', on='MUNICÍPIO')
df_merged
```

Volume de atendimentos por região e município.

Proporção de atendimentos por 100 mil habitantes.

Ranking de municípios com maior e menor volume proporcional de atendimento.

Mapa interativo dos atendimentos SUS.

```python
import plotly.express as px

# =========================
# 1️⃣ Agrupamento
# =========================
atendimentos_por_municipio = (
    df_merged.groupby(['UF', 'MUNICÍPIO'], as_index=False)
    .agg({'PRIMEIRO_NOME': 'count'})
    .rename(columns={'PRIMEIRO_NOME': 'VOLUME_ATENDIMENTOS'})
)

# =========================
# 2️⃣ Gráfico 1 — Volume de atendimentos por UF
# =========================
fig_uf = px.bar(
    atendimentos_por_municipio.groupby('UF', as_index=False)['VOLUME_ATENDIMENTOS'].sum(),
    x='UF',
    y='VOLUME_ATENDIMENTOS',
    color='UF',
    text='VOLUME_ATENDIMENTOS',
    title='📈 Volume de Atendimentos por UF (Região)'
)
fig_uf.update_traces(textposition='outside')
fig_uf.update_layout(
    yaxis_title='Volume de Atendimentos',
    xaxis_title='UF',
    uniformtext_minsize=8,
    uniformtext_mode='hide',
    margin=dict(l=40, r=40, t=60, b=40),
    height=500
)
fig_uf.show()

# =========================
# 3️⃣ Gráfico 2 — Volume de atendimentos por Município
# =========================
fig_mun = px.bar(
    atendimentos_por_municipio.sort_values(by='VOLUME_ATENDIMENTOS', ascending=False),
    x='MUNICÍPIO',
    y='VOLUME_ATENDIMENTOS',
    color='UF',
    text='VOLUME_ATENDIMENTOS',
    title='🏙️ Volume de Atendimentos por Município'
)

# Ajustes de layout e espaçamento
fig_mun.update_traces(textposition='outside', cliponaxis=False)
fig_mun.update_layout(
    xaxis_tickangle=-60,            # Inclina os rótulos do eixo X
    xaxis_tickfont=dict(size=10),   # Tamanho menor da fonte
    yaxis_title='Volume de Atendimentos',
    xaxis_title='Município',
    uniformtext_minsize=8,
    uniformtext_mode='hide',
    bargap=0.25,                    # Espaço entre as colunas
    margin=dict(l=60, r=40, t=80, b=150),  # Aumenta margem inferior
    height=700                      # Altura maior para caber tudo
)
fig_mun.show()
```
